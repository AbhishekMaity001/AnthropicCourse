# Anthropic API and Claude do not store any messages.
- to have a conversation you need to:
    1. Manually maintain a list of messages in your code
    2. Provide that list of messages with each follow up request

# Helper functions

````python
# Helper functions to retain some chat history.

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0, stop_sequences=[]):
    """
    Sending request to claude API.
    """
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences,
    }
    if system is not None:
        params["system"] = system

    # if stop_sequences is not None:
    #     params["stop_sequences"] = stop_sequences
        
    message = client.messages.create(**params)
    print("message", message)
    return message.content[0].text
````

````python
# Make a starting list of messages
messages = []

# Add in the initital user question of "define quantum computing in one sentence"
add_user_message(messages, "Define quantum computing in one sentence")
print(messages)

# calling the claude API
answer = chat(messages)
print(f"1. ===> {answer} \n")
add_assistant_message(messages, answer)

# Add in the users followup question
add_user_message(messages, "Write another sentence")

# Call the chat again with the list of messages to get final answer
answer = chat(messages)
print(f"2. ===> {answer} \n")
````

# Temperature

- Low Temperature - More deterministic output
- High Temperature - More random output

1. Low Temperature (0.0-0.3) - Factual responses, coding assistance, data extraction , Content moderation
2. Medium Temperature (0.4-0.7) - Summarization, Educational content, problem solving, creative writting with constraints
3. High Temperature (0.8-1.0) - Brainstorming, creative writting, marketing content, Joke generation

# Response Streaming

````python
messages = []
add_user_message(messages, "Write a 1 sentence description of a fake database")

stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)
````

````python
for event in stream:
    print(event)
````

## Common events

Event Type  |  Purpose

1. RawMessageStartEvent = A new msg is being sent
2. RawContentBlockStartEvent = Start of a new block, which contains a text message, tool use or other
3. RawContentBlockDeltaEvent = Chunk related to latest content block that was started
4. RawContentBlockStopEvent = The current content block has been completed
5. RawMessageDeltaEvent = The current message is complete
6. RawMessageStopEvent = End of information about the current message






````python
messages = []

add_user_message(messages, "Write a 1 sentence description of a fake database")

with client.messages.stream(
    model=model,
    max_tokens=1000, 
    messages=messages,
) as stream:
    for text in stream.text_stream:
        print(text, end="")
````
